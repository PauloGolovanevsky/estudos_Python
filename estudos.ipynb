# Condicionais

## Condicionais

Condicionais são estruturas de controle que permitem que seu código execute diferentes blocos de instruções dependendo se uma determinada condição é verdadeira ou falsa.

### Como usar:

Em Python, as condicionais são implementadas usando as palavras-chave `if`, `elif` (else if) e `else`.

*   **`if`**: Executa um bloco de código se a condição for verdadeira.
*   **`elif`**: (Opcional) Testa uma nova condição se a condição `if` anterior (e quaisquer `elif`s anteriores) for falsa.
*   **`else`**: (Opcional) Executa um bloco de código se nenhuma das condições `if` ou `elif` anteriores for verdadeira.

#### Sintaxe Básica:

```python
if condicao:
    # Bloco de código a ser executado se a condicao for True
elif outra_condicao:
    # Bloco de código a ser executado se a condicao for False e outra_condicao for True
else:
    # Bloco de código a ser executado se todas as condicoes anteriores forem False
```

### Exemplos Simples e Objetivos:

#### Exemplo 1: `if` simples

```python
idade = 18

if idade >= 18:
    print("Você é maior de idade.")
```

#### Exemplo 2: `if-else`

```python
numero = 10

if numero % 2 == 0:
    print("O número é par.")
else:
    print("O número é ímpar.")
```

#### Exemplo 3: `if-elif-else`

```python
nota = 75

if nota >= 90:
    print("Conceito: A")
elif nota >= 70:
    print("Conceito: B")
elif nota >= 50:
    print("Conceito: C")
else:
    print("Conceito: D")
```

#### Exemplo 4: Múltiplas condições com `and` e `or`

```python
temperatura = 25
esta_chovendo = False

if temperatura > 20 and not esta_chovendo:
    print("Ótimo dia para um passeio!")
elif temperatura < 10 or esta_chovendo:
    print("Melhor ficar em casa.")
else:
    print("Dia normal.")
```

## Exercício 5: Preço de Ingresso de Cinema

Crie um programa que determine o preço de um ingresso de cinema com base na idade do cliente e no dia da semana. Peça ao usuário para digitar a idade e o dia da semana (ex: 'segunda', 'sabado').

Regras de preço:
- **Ingresso Base:** R$ 20
- **Crianças (até 12 anos):** 50% de desconto no ingresso base.
- **Adultos (13 a 64 anos):** Preço base.
- **Idosos (65 anos ou mais):** 30% de desconto no ingresso base.
- **Finais de Semana (sábado, domingo):** Acréscimo de R$ 5 ao preço final para adultos e idosos (crianças não pagam o acréscimo de fim de semana).

Combine operadores lógicos e condicionais aninhadas (`if` dentro de `if`) para calcular o preço final.

````python
idadeCliente = int(input("Digite a sua idade: "))
diaSemana = input("Digite o dia da semana: ").lower()

precoBase = 20

# 1. Define o preço base por idade
if idadeCliente <= 12:
    valorIngresso = precoBase / 2
elif idadeCliente <= 64:
    valorIngresso = precoBase
else:
    valorIngresso = precoBase * 0.7

# 2. Função para verificar fim de semana
def validaFDS(preco, idade, dia):
    if (dia == "sabado" or dia == "domingo") and idade > 12:
        preco += 5
    return preco

# 3. Aplica regra de fim de semana
valorFinal = validaFDS(valorIngresso, idadeCliente, diaSemana)

# 4. Exibe resultado
print("O valor do ingresso é:", valorFinal)
````

# Laços de Repetição

Laços de repetição (ou loops) são estruturas que permitem executar um bloco de código repetidamente, com base em uma condição ou para cada item em uma sequência.

### Como usar:

Em Python, os principais tipos de laços de repetição são `for` e `while`.

*   **`for`**: Usado para iterar sobre uma sequência (como uma lista, tupla, string, ou range) ou outros objetos iteráveis. Ele executa o bloco de código para cada item na sequência.
*   **`while`**: Executa um bloco de código repetidamente enquanto uma condição especificada for verdadeira. O loop continua até que a condição se torne falsa.

#### Sintaxe Básica:

```python
# Laço for
for item in sequencia:
    # Bloco de código a ser executado para cada item

# Laço while
while condicao:
    # Bloco de código a ser executado enquanto a condicao for True
    # É importante ter algo dentro do loop que eventualmente torne a condicao False
```

### Exemplos Simples e Objetivos:

#### Exemplo 1: `for` com lista

````python
frutas = ['maçã', 'banana', 'cereja']
for fruta in frutas:
    print(f'Eu gosto de {fruta}.')
````

#### Exemplo 2: `while` simples (contagem)

````python
contador = 0
while contador < 5:
    print(f'Contador é {contador}')
    contador += 1
````

#### Exemplo 3: `for` com `range()`

````python
print('Contando de 0 a 4:')
for i in range(5):
    print(i)

print('\nContando de 2 a 7:')
for i in range(2, 8):
    print(i)

print('\nContando de 10 a 0 (de 2 em 2):')
for i in range(10, -1, -2):
    print(i)
````

#### Exemplo 4: `for` com `break` e `continue`

*   **`break`**: Sai completamente do loop.
*   **`continue`**: Pula a iteração atual e vai para a próxima.

````python
print('Exemplo com break:')
for i in range(10):
    if i == 5:
        print('Encontrei o 5, saindo do loop.')
        break
    print(i)

print('\nExemplo com continue:')
for i in range(5):
    if i == 2:
        print('Pulando o 2.')
        continue
    print(i)
````

# Estrtura de Dados

Estrutura de dados

Estruturas de dados são formas organizadas de armazenar e gerenciar dados, permitindo acesso e modificação eficientes. Em Python, existem várias estruturas de dados embutidas, cada uma com suas características e usos específicos.

### Como usar:

As estruturas de dados mais comuns em Python são:

*   **Listas (`list`)**: Coleções ordenadas e mutáveis que podem armazenar itens de diferentes tipos. São definidas usando colchetes `[]`.
*   **Tuplas (`tuple`)**: Coleções ordenadas e imutáveis (não podem ser alteradas após a criação). São definidas usando parênteses `()`.
*   **Dicionários (`dict`)**: Coleções não ordenadas de pares chave-valor, onde cada chave é única. São definidos usando chaves `{}`.
*   **Conjuntos (`set`)**: Coleções não ordenadas de itens únicos. São definidos usando chaves `{}` ou a função `set()`.

#### Sintaxe Básica:

```python
# Lista
minha_lista = [1, 'dois', 3.0, True]

# Tupla
minha_tupla = (10, 'vinte', False)

# Dicionário
meu_dicionario = {'nome': 'Alice', 'idade': 30, 'cidade': 'São Paulo'}

# Conjunto
meu_conjunto = {1, 2, 3, 2, 1} # Automaticamente remove duplicatas
```

### Exemplos Simples e Objetivos:

#### Exemplo 1: Listas

````python
# Criação e acesso
minha_lista = ['uva', 'maçã', 'banana', 'laranja']
print('Lista original:', minha_lista)
print('Primeiro elemento:', minha_lista[0])
print('Último elemento:', minha_lista[-1])

# Modificação
minha_lista[1] = 'pera'
print('Lista modificada:', minha_lista)

# Adicionar elementos
minha_lista.append('abacaxi')
minha_lista.insert(1, 'morango')
print('Lista com novos elementos:', minha_lista)

# Remover elementos
minha_lista.remove('banana')
item_removido = minha_lista.pop(0) # Remove e retorna o item no índice 0
print('Item removido com pop:', item_removido)
print('Lista após remoções:', minha_lista)

# Fatiamento (Slicing)
sub_lista = minha_lista[1:3]
print('Sub-lista (fatiamento):', sub_lista)
````

#### Exemplo 2: Tuplas

````python
# Criação e acesso
minha_tupla = (1, 2, 3, 'quatro', 5.0)
print('Tupla original:', minha_tupla)
print('Segundo elemento:', minha_tupla[1])

# Imutabilidade (tentar modificar causaria um erro)
# minha_tupla[0] = 99 # Isso geraria um TypeError

# Contagem e índice
print('Número de vezes que 2 aparece:', minha_tupla.count(2))
print('Índice do elemento "quatro":', minha_tupla.index('quatro'))
````

#### Exemplo 3: Dicionários

````python
# Criação
meu_dicionario = {
    'nome': 'Carlos',
    'idade': 25,
    'cidade': 'Rio de Janeiro'
}
print('Dicionário original:', meu_dicionario)

# Acesso a valores
print('Nome:', meu_dicionario['nome'])
print('Idade:', meu_dicionario.get('idade')) # Usar .get() é mais seguro

# Adicionar/Modificar pares chave-valor
meu_dicionario['profissão'] = 'Engenheiro'
meu_dicionario['idade'] = 26
print('Dicionário após modificações:', meu_dicionario)

# Remover um par
valor_removido = meu_dicionario.pop('cidade')
print('Valor da cidade removida:', valor_removido)
print('Dicionário após remover cidade:', meu_dicionario)

# Iterar sobre o dicionário
print('\nChaves do dicionário:')
for chave in meu_dicionario.keys():
    print(chave)

print('\nValores do dicionário:')
for valor in meu_dicionario.values():
    print(valor)

print('\nPares chave-valor do dicionário:')
for chave, valor in meu_dicionario.items():
    print(f'{chave}: {valor}')
````

#### Exemplo 4: Conjuntos (Sets)

````python
# Criação (duplicatas são automaticamente removidas)
meu_conjunto = {1, 2, 3, 3, 4, 5, 1}
print('Conjunto original:', meu_conjunto)

# Adicionar elementos
meu_conjunto.add(6)
print('Conjunto após adicionar 6:', meu_conjunto)

# Remover elementos
meu_conjunto.remove(2) # Gera erro se o elemento não existir
print('Conjunto após remover 2:', meu_conjunto)
meu_conjunto.discard(10) # Não gera erro se o elemento não existir
print('Conjunto após tentar remover 10 (inexistente):', meu_conjunto)

# Operações de conjunto
conjunto_a = {1, 2, 3, 4}
conjunto_b = {3, 4, 5, 6}

uniao = conjunto_a.union(conjunto_b)
intersecao = conjunto_a.intersection(conjunto_b)
diferenca_a_b = conjunto_a.difference(conjunto_b)
diferenca_b_a = conjunto_b.difference(conjunto_a)

print('\nConjunto A:', conjunto_a)
print('Conjunto B:', conjunto_b)
print('União (A | B):', uniao)
print('Interseção (A & B):', intersecao)
print('Diferença (A - B):', diferenca_a_b)
print('Diferença (B - A):', diferenca_b_a)
````
